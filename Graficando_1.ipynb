```python
import pandas as pd
import numpy as np
url='https://raw.githubusercontent.com/Diehgon/Curso-coderhouse-DS/main/water_potability.csv'
df= pd.read_csv(url, sep= ',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
resumen= df.describe().round(2)
resumen
```

```python
# contar el numero de valores que son 0 por columnas
nun_missing = (df == 0).sum()
print(nun_missing)
```

```python
# replace '0' values with 'nan'
df[['ph', 'Sulfate', 'Trihalomethanes']]= df[['ph', 'Sulfate', 'Trihalomethanes']].replace(0, np.nan)
# count the number of nan values in each column
print(df.isnull().sum())
```

```python
df.loc[df.ph.isna()]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Primer grafico con seaborn
sns.histplot(data=df, x="Potability",bins=5)
```

```python
df_n=df[['Hardness','Conductivity','Potability']].groupby(by=['Potability']).count().reset_index()
df_n
```

```python
#Segundo grafico con seaborn
plt.figure(figsize=(10,6))
sns.barplot(y='Hardness',x='Conductivity',hue='Potability',data=df_n);
```

```python
#tercer grafico
sns.displot(data=df, x='Turbidity', hue='Potability', kind="kde")
plt.show()
```
